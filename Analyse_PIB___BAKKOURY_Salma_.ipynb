```python
"""
Analyse Complète du PIB Mondial 2024
Auteur: Analyse Économique
Date: Octobre 2024
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuration
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# ===================================
# DONNÉES PIB 2024
# ===================================

donnees_pib = {
    'Pays': [
        'États-Unis', 'Chine', 'Allemagne', 'Japon', 'Inde',
        'Royaume-Uni', 'France', 'Italie', 'Canada', 'Brésil',
        'Russie', 'Corée du Sud', 'Mexique', 'Australie', 'Espagne',
        'Indonésie', 'Turquie', 'Pays-Bas', 'Arabie Saoudite', 'Suisse'
    ],
    'PIB_2024_Milliards_USD': [
        29167.78, 21643.00, 4456.00, 4365.00, 3889.13,
        3587.55, 3174.10, 2376.51, 2214.80, 2188.42,
        2100.00, 1800.00, 1700.00, 1680.00, 1580.00,
        1400.00, 1350.00, 1120.00, 1100.00, 950.00
    ],
    'Croissance_2024_Pourcent': [
        2.1, 3.0, 1.8, 1.0, 7.0,
        4.1, 2.6, 0.8, 3.1, 2.5,
        -2.2, 2.3, 3.2, 2.8, 2.4,
        5.1, 4.5, 1.5, 3.8, 1.2
    ],
    'Region': [
        'Amérique du Nord', 'Asie', 'Europe', 'Asie', 'Asie',
        'Europe', 'Europe', 'Europe', 'Amérique du Nord', 'Amérique du Sud',
        'Europe', 'Asie', 'Amérique du Nord', 'Océanie', 'Europe',
        'Asie', 'Europe', 'Europe', 'Moyen-Orient', 'Europe'
    ],
    'Population_Millions': [
        340, 1425, 84, 125, 1430,
        68, 68, 59, 40, 216,
        144, 52, 128, 26, 48,
        277, 85, 18, 36, 9
    ],
    'PIB_2023_Milliards_USD': [
        28550, 21016, 4370, 4321, 3637,
        3445, 3095, 2358, 2147, 2135,
        2148, 1759, 1647, 1635, 1543,
        1330, 1293, 1103, 1061, 938
    ]
}

# Création du DataFrame
df = pd.DataFrame(donnees_pib)

# Calculs supplémentaires
df['PIB_par_Habitant_USD'] = (df['PIB_2024_Milliards_USD'] * 1000) / df['Population_Millions']
df['Variation_PIB_2023_2024_Milliards'] = df['PIB_2024_Milliards_USD'] - df['PIB_2023_Milliards_USD']
df['Part_PIB_Mondial_Pourcent'] = (df['PIB_2024_Milliards_USD'] / df['PIB_2024_Milliards_USD'].sum()) * 100

# ===================================
# AFFICHAGE DES DONNÉES
# ===================================

print("=" * 100)
print("ANALYSE COMPLÈTE DU PIB MONDIAL 2024 - TOP 20 ÉCONOMIES".center(100))
print("=" * 100)
print()

# Tableau principal
print("📊 TABLEAU COMPLET DES DONNÉES")
print("-" * 100)
df_affichage = df[['Pays', 'PIB_2024_Milliards_USD', 'Croissance_2024_Pourcent',
                    'PIB_par_Habitant_USD', 'Region', 'Part_PIB_Mondial_Pourcent']].copy()
df_affichage['PIB_2024_Milliards_USD'] = df_affichage['PIB_2024_Milliards_USD'].apply(lambda x: f"{x:,.0f}")
df_affichage['PIB_par_Habitant_USD'] = df_affichage['PIB_par_Habitant_USD'].apply(lambda x: f"{x:,.0f}")
df_affichage['Croissance_2024_Pourcent'] = df_affichage['Croissance_2024_Pourcent'].apply(lambda x: f"{x:+.1f}%")
df_affichage['Part_PIB_Mondial_Pourcent'] = df_affichage['Part_PIB_Mondial_Pourcent'].apply(lambda x: f"{x:.2f}%")
print(df_affichage.to_string(index=False))
print()

# Statistiques clés
print("=" * 100)
print("📈 STATISTIQUES CLÉS")
print("=" * 100)
print(f"💰 PIB Total (Top 20):              {df['PIB_2024_Milliards_USD'].sum():>15,.0f} Milliards USD")
print(f"📊 PIB Mondial estimé 2024:         {103000:>15,.0f} Milliards USD")
print(f"📈 Croissance moyenne (Top 20):     {df['Croissance_2024_Pourcent'].mean():>15.2f}%")
print(f"🌍 PIB moyen par pays:              {df['PIB_2024_Milliards_USD'].mean():>15,.0f} Milliards USD")
print(f"👥 PIB/hab moyen (Top 20):          {df['PIB_par_Habitant_USD'].mean():>15,.0f} USD")
print()
print(f"🥇 Plus grand PIB:                  {df.iloc[0]['Pays']} ({df.iloc[0]['PIB_2024_Milliards_USD']:,.0f}B USD)")
print(f"🚀 Plus forte croissance:           {df.loc[df['Croissance_2024_Pourcent'].idxmax(), 'Pays']} ({df['Croissance_2024_Pourcent'].max():.1f}%)")
print(f"💎 Plus riche par habitant:         {df.loc[df['PIB_par_Habitant_USD'].idxmax(), 'Pays']} ({df['PIB_par_Habitant_USD'].max():,.0f} USD)")
print(f"📉 Seule décroissance:               {df.loc[df['Croissance_2024_Pourcent'].idxmin(), 'Pays']} ({df['Croissance_2024_Pourcent'].min():.1f}%)")
print()

# Analyse par région
print("=" * 100)
print("🌍 ANALYSE PAR RÉGION")
print("=" * 100)
analyse_region = df.groupby('Region').agg({
    'PIB_2024_Milliards_USD': ['sum', 'mean', 'count'],
    'Croissance_2024_Pourcent': 'mean',
    'Population_Millions': 'sum'
}).round(2)
analyse_region.columns = ['PIB Total (B$)', 'PIB Moyen (B$)', 'Nb Pays', 'Croissance Moy (%)', 'Population Tot (M)']
analyse_region = analyse_region.sort_values('PIB Total (B$)', ascending=False)
print(analyse_region)
print()

# Top 5 par catégorie
print("=" * 100)
print("🏆 TOP 5 PAR CATÉGORIE")
print("=" * 100)
print("\n💰 TOP 5 - PIB TOTAL:")
print(df.nlargest(5, 'PIB_2024_Milliards_USD')[['Pays', 'PIB_2024_Milliards_USD', 'Region']].to_string(index=False))
print("\n📈 TOP 5 - CROISSANCE:")
print(df.nlargest(5, 'Croissance_2024_Pourcent')[['Pays', 'Croissance_2024_Pourcent', 'Region']].to_string(index=False))
print("\n💎 TOP 5 - PIB PAR HABITANT:")
print(df.nlargest(5, 'PIB_par_Habitant_USD')[['Pays', 'PIB_par_Habitant_USD', 'Population_Millions']].to_string(index=False))
print("\n📊 TOP 5 - VARIATION 2023-2024:")
print(df.nlargest(5, 'Variation_PIB_2023_2024_Milliards')[['Pays', 'Variation_PIB_2023_2024_Milliards', 'Croissance_2024_Pourcent']].to_string(index=False))
print()

# ===================================
# VISUALISATIONS
# ===================================

print("=" * 100)
print("📊 GÉNÉRATION DES GRAPHIQUES...")
print("=" * 100)

# Création de la figure
fig = plt.figure(figsize=(20, 12))
fig.suptitle('ANALYSE COMPLÈTE DU PIB MONDIAL 2024 - TOP 20 ÉCONOMIES',
             fontsize=20, fontweight='bold', y=0.98)

# 1. Top 20 PIB
ax1 = plt.subplot(3, 3, (1, 3))
couleurs = plt.cm.viridis(np.linspace(0.2, 0.9, len(df)))
bars = ax1.barh(df['Pays'], df['PIB_2024_Milliards_USD'], color=couleurs, edgecolor='black', linewidth=0.5)
ax1.set_xlabel('PIB 2024 (Milliards USD)', fontweight='bold', fontsize=11)
ax1.set_title('🌍 Classement des 20 Premières Économies Mondiales', fontsize=14, fontweight='bold', pad=15)
ax1.invert_yaxis()
for i, (val, croiss) in enumerate(zip(df['PIB_2024_Milliards_USD'], df['Croissance_2024_Pourcent'])):
    couleur_txt = 'green' if croiss > 0 else 'red'
    ax1.text(val + 600, i, f'{val:,.0f}B ({croiss:+.1f}%)',
             va='center', fontsize=8, color=couleur_txt, fontweight='bold')
ax1.grid(axis='x', alpha=0.3, linestyle='--')

# 2. Répartition par région (Pie chart)
ax2 = plt.subplot(3, 3, 4)
pib_region = df.groupby('Region')['PIB_2024_Milliards_USD'].sum().sort_values(ascending=False)
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']
wedges, texts, autotexts = ax2.pie(pib_region, labels=pib_region.index, autopct='%1.1f%%',
                                     startangle=90, colors=colors, explode=[0.05]*len(pib_region))
ax2.set_title('📊 Répartition du PIB par Région', fontsize=12, fontweight='bold', pad=10)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# 3. Taux de croissance
ax3 = plt.subplot(3, 3, (5, 6))
df_tri = df.sort_values('Croissance_2024_Pourcent', ascending=True)
couleurs_croiss = ['#d62728' if x < 0 else '#2ca02c' if x > 4 else '#ff7f0e'
                   for x in df_tri['Croissance_2024_Pourcent']]
ax3.barh(df_tri['Pays'], df_tri['Croissance_2024_Pourcent'], color=couleurs_croiss, alpha=0.7)
ax3.axvline(x=0, color='black', linestyle='-', linewidth=2)
ax3.axvline(x=2.9, color='blue', linestyle='--', linewidth=1.5, alpha=0.6, label='Moyenne (~2.9%)')
ax3.set_xlabel('Taux de Croissance 2024 (%)', fontweight='bold', fontsize=11)
ax3.set_title('📈 Taux de Croissance du PIB par Pays', fontsize=12, fontweight='bold', pad=10)
ax3.legend(loc='lower right')
ax3.grid(axis='x', alpha=0.3)

# 4. PIB par habitant (Top 15)
ax4 = plt.subplot(3, 3, 7)
df_capita = df.nlargest(15, 'PIB_par_Habitant_USD')
couleurs_cap = plt.cm.plasma(np.linspace(0.2, 0.9, len(df_capita)))
bars4 = ax4.bar(range(len(df_capita)), df_capita['PIB_par_Habitant_USD'],
                color=couleurs_cap, alpha=0.8, edgecolor='black', linewidth=0.5)
ax4.set_xticks(range(len(df_capita)))
ax4.set_xticklabels(df_capita['Pays'], rotation=45, ha='right', fontsize=9)
ax4.set_ylabel('PIB par Habitant (USD)', fontweight='bold', fontsize=10)
ax4.set_title('💰 TOP 15 - PIB par Habitant', fontsize=12, fontweight='bold', pad=10)
for i, val in enumerate(df_capita['PIB_par_Habitant_USD']):
    ax4.text(i, val + 2000, f'{val:,.0f}', ha='center', va='bottom', fontsize=7, fontweight='bold')
ax4.grid(axis='y', alpha=0.3)

# 5. Scatter PIB vs Population
ax5 = plt.subplot(3, 3, 8)
scatter = ax5.scatter(df['Population_Millions'], df['PIB_2024_Milliards_USD'],
                      s=df['Croissance_2024_Pourcent'].abs()*80,
                      c=df['Croissance_2024_Pourcent'],
                      cmap='RdYlGn', alpha=0.7, edgecolors='black', linewidth=1.5)
for i in range(min(10, len(df))):
    ax5.annotate(df['Pays'].iloc[i],
                (df['Population_Millions'].iloc[i], df['PIB_2024_Milliards_USD'].iloc[i]),
                xytext=(5, 5), textcoords='offset points', fontsize=7, fontweight='bold')
ax5.set_xlabel('Population (Millions)', fontweight='bold', fontsize=10)
ax5.set_ylabel('PIB (Milliards USD)', fontweight='bold', fontsize=10)
ax5.set_title('🔍 PIB vs Population\n(Taille bulle = Croissance)', fontsize=11, fontweight='bold', pad=10)
cbar = plt.colorbar(scatter, ax=ax5)
cbar.set_label('Croissance (%)', fontweight='bold')
ax5.grid(alpha=0.3)

# 6. Évolution 2023-2024
ax6 = plt.subplot(3, 3, 9)
df_evol = df.nlargest(10, 'Variation_PIB_2023_2024_Milliards')
x = np.arange(len(df_evol))
width = 0.35
bars1 = ax6.bar(x - width/2, df_evol['PIB_2023_Milliards_USD'], width, label='2023', alpha=0.8, color='#1f77b4')
bars2 = ax6.bar(x + width/2, df_evol['PIB_2024_Milliards_USD'], width, label='2024', alpha=0.8, color='#ff7f0e')
ax6.set_xlabel('Pays', fontweight='bold', fontsize=10)
ax6.set_ylabel('PIB (Milliards USD)', fontweight='bold', fontsize=10)
ax6.set_title('📊 Évolution 2023-2024 (Top 10)', fontsize=12, fontweight='bold', pad=10)
ax6.set_xticks(x)
ax6.set_xticklabels(df_evol['Pays'], rotation=45, ha='right', fontsize=8)
ax6.legend()
ax6.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('analyse_pib_mondial_2024.png', dpi=300, bbox_inches='tight', facecolor='white')
print("\n✅ Graphique sauvegardé: 'analyse_pib_mondial_2024.png'")
plt.show()

# ===================================
# EXPORT DES DONNÉES
# ===================================

print("\n" + "=" * 100)
print("💾 EXPORT DES DONNÉES")
print("=" * 100)

# Export CSV
df.to_csv('pib_mondial_2024.csv', index=False, encoding='utf-8-sig')
print("✅ Fichier CSV exporté: 'pib_mondial_2024.csv'")

# Export Excel
try:
    df.to_excel('pib_mondial_2024.xlsx', index=False, sheet_name='PIB 2024')
    print("✅ Fichier Excel exporté: 'pib_mondial_2024.xlsx'")
except:
    print("⚠️  Pour exporter en Excel, installez: pip install openpyxl")

print("\n" + "=" * 100)
print("✅ ANALYSE TERMINÉE AVEC SUCCÈS!")
print("=" * 100)
```
